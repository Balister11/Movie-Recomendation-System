```python
# Add the data completely so that we can apply cosine similarity easily:

complete_data = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
# converting into numbers: text to numbers : tfidf [Term Frequency and Inverse document frequency]

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
cv = TfidfVectorizer()
```

```python
features_extracted = cv.fit_transform(complete_data)
```

```python
print(features_extracted)
```

```python
# cosine similarity concept:
```

```python
# so for implemeting cosine similarity let's import some stuff

import difflib
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity_ = cosine_similarity(features_extracted)
cosine_similarity_
```

```python
cosine_similarity_.shape
```

```python
movie_name = 'iron man'
```

```python
# Lets find out first recomendation:



all_names_of_movies = df['title'].tolist()
all_names_of_movies
```

```python
final_pred = difflib.get_close_matches(movie_name , all_names_of_movies)
final_pred
```

```python
# finding it in more clear way and get nore recomendations:

check = input('Enter a movie name : ')  # iron man 

# find the index of the movie title what ur giving 

index_of_movie = df[df['title'] == check]['index'].values[0]

print('index value :' , index_of_movie)

scores = list(enumerate(cosine_similarity_[index_of_movie]))

#print(scores)

# lets sort the data and select a limit for recomendation:

sorted_data = sorted(scores , key = lambda x:x[1] , reverse=True)

#print(sorted_data)


# Movie Recomendation:  by collecting title 

count = 1

for i in sorted_data:
    index_ = i[0]
    title_ = df[df.index == index_]['title'].values[0]
    if count < 30:
        print(count , '-----' , title_)
        
        count+=1
```
